```python
data['target'] = data['target'].apply(lambda x: int(x.split('_')[1]))
```

```python
data.isnull().sum()
```

EDA

Outliers identifying and removal

```python
plt.figure(figsize=(15,8))
ax = sns.boxplot(data=data, orient='h')
```

```python
Q1 = data.quantile(0.1)
Q3 = data.quantile(0.9)
IQR = Q3 - Q1

data = data[~((data < (Q1 - 1.5 * IQR)) |(data > (Q3 + 1.5 * IQR))).any(axis=1)]
```

```python
ax = sns.boxplot(data=data, orient='h')
```

```python
for col in data.columns:
    print(col)
    values = data[col].value_counts()
    plt.bar(x=values.index, height=values)
    plt.show()
   
```

```python
data.drop([col for col in data.columns if len(data[col].unique()) == 1], inplace=True, axis=1)
```

```python
data.shape
```

Train|Test

```python
data_train, data_test, target_train, target_test = train_test_split(data.drop('target', axis=1), data['target'])
```

```python
data_test.index
```

### Prediction

1. DecisionTree
2. RandomForest
3. Log

```python
tree_model = tree.DecisionTreeClassifier()

kf = KFold(n_splits=3)

for train, test in kf.split(data_train):
    X_train, X_test, y_train, y_test  = data_train[data_train.index.isin(train)], data_train[data_train.index.isin(test)], target_train[data_train.index.isin(train)],  target_train[data_train.index.isin(test)]

    tree_model.fit(X_train, y_train)
    print('accuracy:', mse(tree_model.predict(X_test), y_test))
```
